```python
text = """JAKARTA, KOMPAS.com - Komisi Pemberantasan Korupsi (KPK) mengatakan, nama pengusaha rokok Khairul Umam alias Haji Her ada tercantum dalam dokumen yang ditemukan penyidik saat menggeledah kantor Direktorat Jenderal Bea Cukai (DJBC)."""
```

```python
clean_text = re.sub(r'[^a-zA-Z\s]', ' ', text)
clean_text = re.sub(r'\s+', ' ', clean_text)

print("1. Cleaning:\n", clean_text)
```

```python
case_folding = clean_text.lower()

print("\n2. Case Folding:\n", case_folding)
```

```python
tokens = word_tokenize(case_folding)

print("\n3. Tokenization:\n", tokens)
```

```python
normalization_dict = {
    "kompascom": "kompas",
    "dokumendokumen": "dokumen",
    "namanama": "nama"
}

normalized = [normalization_dict.get(word, word) for word in tokens]

print("\n4. Normalization:\n", normalized)
```

```python
stop_words = set(stopwords.words('indonesian'))

filtered = [word for word in normalized if word not in stop_words]

print("\n5. Stopword Removal:\n", filtered)
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

stemmed = [stemmer.stem(word) for word in filtered]

print("\n6. Stemming:\n", stemmed)
```

```python
lemmatized = stemmed
print("\n7. Lemmatization:\n", lemmatized)
```
